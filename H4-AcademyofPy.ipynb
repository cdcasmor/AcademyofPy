```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
students_df = pd.read_csv("students_complete.csv")
schools_df = pd.read_csv("schools_complete.csv")
```

```python
#Finding the total number of schools
schools_count = int(schools_df.shape[0]-1)
print (schools_count)
```

```python
#Finding the total number of students
students_count = int(students_df.shape[0]-1)
print (students_count)
```

```python
#finding the total budget
total_budget = int(schools_df['budget'].sum())
print(total_budget)
```

```python
#Finding the avg math score
avg_math = students_df['math_score'].mean()
print(avg_math)
```

```python
#Finding the avg reading score 
avg_reading = students_df['reading_score'].mean()
print(avg_reading)
```

```python
#Find the passing percentages
students_passingmath = students_df.loc[(students_df["math_score"] >= 70)]
students_passingreading = students_df.loc[(students_df["reading_score"] >= 70)]
passing_count_m = int(students_passingmath.shape[0]-1)
passing_count_r = int(students_passingreading.shape[0]-1)
percent_passing_m = ((passing_count_m/students_count)*100)
percent_passing_r = ((passing_count_r/students_count)*100)
percent_passing = ((((passing_count_m/students_count)*100) + ((passing_count_r/students_count)*100))/2)
print(percent_passing_m)
print(percent_passing_r)
print(percent_passing)
```

```python
District_df = pd.DataFrame(
            { 'Total Schools': [schools_count],
              'Total Students': [students_count],
              'Total Budget': [total_budget],
              'Average Math Score' : [avg_math],
              'Average Reading Score': [avg_reading],
              '% Passing Math' : [percent_passing_m],
              '% Passing Reading'  : [percent_passing_r],
              'Overall Passing Rate' : [percent_passing]})
```
